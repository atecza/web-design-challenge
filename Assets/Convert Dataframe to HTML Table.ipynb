```python
import pandas as pd
```

```python
weather_df = pd.read_csv('Resources/weather_data.csv')
```

```python
weather_df.reset_index(drop=True, inplace=True)
```

```python
weather_df = weather_df.iloc[:,1:]
```

```python
weather_df.head()
```

```python
weather_html = weather_df.to_html(index = False, justify = 'left', classes= ['table table-striped table-bordered table-hover'] )

#write html to file
text_file = open("weather_table.html", "w")
text_file.write(weather_html)
text_file.close()
```
